```python
def delete_files(path_png, path_txt, f_list):
    for f in f_list:
        txt_name = f[-4:]+'.txt'
        
        if os.path.exists(os.path.join(path_png, f)):
            print(f)
            os.remove(os.path.join(path_png, f))
        
        if os.path.exists(os.path.join(path_txt, txt_name)):
            #print(f)
            os.remove(os.path.join(path_txt, txt_name))
    return
```

```python
# Clean Training data Transfer 
for f in os.listdir(train_path):
    hashmap = f.split('_')[1][:-4]
    #print(hashmap)
    negsamp_pp = 'NEGSAMP_'+hashmap+'_pp.png'
    negsamp_mp = 'NEGSAMP_'+hashmap+'_mp.png'
    negsamp_pm = 'NEGSAMP_'+hashmap+'_pm.png'
    negsamp_mm = 'NEGSAMP_'+hashmap+'_mm.png'
    #print(negsamp_pp)
    basefile = os.path.join(train_path, f)
    
    neg_pp = os.path.join(base_train_path, negsamp_pp)
    neg_mp = os.path.join(base_train_path, negsamp_mp)
    neg_pm = os.path.join(base_train_path, negsamp_pm)
    neg_mm = os.path.join(base_train_path, negsamp_mm)
    
    try: 
        copy2(basefile, new_train_pos)
    except:
        pass
    
    try:
        copy2(neg_pp, new_train_neg)
        copy2(neg_pm, new_train_neg)
        copy2(neg_mp, new_train_neg)
        copy2(neg_mm, new_train_neg)
    
    except:
        pass
```
