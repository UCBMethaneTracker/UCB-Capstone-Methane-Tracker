```python
import os
import pandas as pd
import pickle
```

```python
path = 'D:/MIDS/CAPSTONE/UCB-Capstone-Methane-Tracker'
hd1 = 'full_training_datadict.pkl'
hd2 = 'full_test_datadict.pkl'
hd3 = 'training_datadict.pkl'
hd4 = 'test_datadict.pkl'
```

```python
with open(os.path.join(path, hd1), "rb") as hashed_l:   
            full_train = pickle.load(hashed_l)

with open(os.path.join(path, hd2), "rb") as hashed_l:   
            full_test = pickle.load(hashed_l)
        
with open(os.path.join(path, hd3), "rb") as hashed_l:   
            train = pickle.load(hashed_l)
        
with open(os.path.join(path, hd4), "rb") as hashed_l:   
            test = pickle.load(hashed_l)
```

```python
results_train, results_test = [], []


for k, v in full_train.items():
    if train.get(k, None) is None:
        results_train.append('BASE_'+str(k)+'.png')
        results_train.append('NEGSAMP_'+str(k)+'_pp.png')
        results_train.append('NEGSAMP_'+str(k)+'_pm.png')
        results_train.append('NEGSAMP_'+str(k)+'_mp.png')
        results_train.append('NEGSAMP_'+str(k)+'_mm.png')

for k, v in full_test.items():
    if test.get(k, None) is None:
        results_test.append('BASE_'+str(k)+'.png')
        results_test.append('NEGSAMP_'+str(k)+'_pp.png')
        results_test.append('NEGSAMP_'+str(k)+'_pm.png')
        results_test.append('NEGSAMP_'+str(k)+'_mp.png')
        results_test.append('NEGSAMP_'+str(k)+'_mm.png')
```

```python
len(results_train) == 5*(len(full_train) - len(train))
```

```python
len(results_test) == 5*(len(full_test) - len(test))
```

```python
df_train = pd.DataFrame()
df_test = pd.DataFrame()
```

```python
df_train['new_filenames'] = results_train
df_test['new_filenames'] = results_test
```

```python
df_train.to_csv('new_training_files.csv', index=False)
df_test.to_csv('new_test_files.csv', index=False)
```
